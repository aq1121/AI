### 合并表

```python
data = pd.merge(ratings,movies)
```

```python
data.head(10)
```

## 电影平均分

```python
avarage = data.groupby('movieId').agg({'rating':np.mean})
```

```python
avarage['title'] = data['title'].groupby(data['movieId']).first()
```

```python
avarage['count'] =  data['title'].groupby(data['movieId']).count()
```
